1. Проверка наличия элемента в массиве

```python
def contains(arr, x):
    for el in arr:
        if el == x:
            return True
    return False
```

2. Поиск второго максимального элемента

```python
def second_max(arr):
    max1 = max2 = float('-inf')

    for x in arr:
        if x > max1:
            max2 = max1
            max1 = x
        elif x > max2 and x != max1:
            max2 = x

    return max2
```

3. Бинарный поиск

```python
def binary_search(arr, x):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == x:
            return True
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1

    return False
```

4. Построение таблицы умножения

```python
def multiplication_table(n):
    table = []
    for i in range(1, n + 1):
        row = []
        for j in range(1, n + 1):
            row.append(i * j)
        table.append(row)
    return table
```

 5. Функция измерения времени

```python
import time

def measure_time(func, *args):
    start = time.perf_counter()
    func(*args)
    end = time.perf_counter()
    return end - start
```
